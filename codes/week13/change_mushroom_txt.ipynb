```python
import numpy as np
mush = np.asarray(mushroom)
mush
```

```python
for i in range(len(mush)):
    for j in range(1,len(mush[0])):
        mush[i][j] = ord(mush[i][j])-ord('a')
```

```python
for i in range(len(mush)):
    if mush[i][0] == 'e': mush[i][0]= 0
    else: mush[i][0] = 1
```

```python
(pd.DataFrame(mush)).to_csv('mushroom2.txt', index = False, sep= ' ')
```
